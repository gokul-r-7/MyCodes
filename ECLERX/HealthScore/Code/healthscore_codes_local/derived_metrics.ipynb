```python
def calculate_metrics(df, new_row_display_name, numerator_display_name, denominator_display_name, operation):
    """
    Calculate and append a new row based on the given display names for the numerator and denominator, 
    and an operation ('multiply' or 'divide') to be applied.

    Args:
        df (pd.DataFrame): The input DataFrame containing the original data.
        new_row_display_name (str): The name of the new row (metric) to generate (e.g., 'Total Time spent').
        numerator_display_name (str): The display name used as the numerator for the formula.
        denominator_display_name (str): The display name used as the denominator for the formula.
        operation (str): The operation to perform - 'multiply' or 'divide'.

    Returns:
        pd.DataFrame: The updated DataFrame with the calculated metric appended.
    """
    os_types = df['operating_system_type'].unique()  # Get unique OS types (ios, android)
    new_rows = []

    # Loop over each operating system type
    for os in os_types:
        os_data = df[df['operating_system_type'] == os]  # Filter data for the specific OS
        
        # Extract the relevant values based on display_names
        numerator_values = os_data[os_data['display_names'] == numerator_display_name].iloc[:, 2:].values
        denominator_values = os_data[os_data['display_names'] == denominator_display_name].iloc[:, 2:].values
        
        # Ensure the operation is either 'multiply' or 'divide'
        if operation == 'multiply':
            result = numerator_values * denominator_values
        elif operation == 'divide':
            result = numerator_values / denominator_values
        else:
            raise ValueError("Operation must be 'multiply' or 'divide'")

        # Initialize a dictionary to hold the values for the new row
        new_row_values = {df.columns[i]: result[0][i-2] for i in range(2, len(df.columns))}

        # Append the calculated values as a new row in the correct format
        new_rows.append({
            'display_names': new_row_display_name,
            'operating_system_type': os,
            **new_row_values
        })
    
    # Create a DataFrame for the new rows
    new_df = pd.DataFrame(new_rows)

    # Concatenate the new rows with the original DataFrame
    df = pd.concat([df, new_df], ignore_index=True)

    return df
```
