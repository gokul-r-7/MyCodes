```python
import pandas as pd

# Data for Devices Page Visits
dates = [
    "2024-08-20", "2024-08-21", "2024-08-22", 
    "2024-08-23", "2024-08-24", "2024-08-25", 
    "2024-08-26", "2024-08-27"
]

visits = [
    3321, 4459, 3690, 
    3914, 3351, 2984, 
    3420, 3403
]

percentages_data = {
    "Saw a Result": [5.1, 3.8, 3.6, 2.8, 3.0, 3.7, 3.2, 3.24],
    "1st common reason (Unknown)": [67.6, 77.6, 78.8, 74.6, 72.3, 81.1, 77.6, 79.4],
    "2nd common reason (Disconnected)": [8.6, 12.1, 11.8, 14.1, 10.8, 9.5, 7.5, 13.2],
}

# Prepare the Operating System Type
os_types = ['a' if i % 2 == 0 else 'g' for i in range(len(dates))]

# Create the DataFrame
data = {
    "Dates": dates,
    "Operating System Type": os_types,
    "Metric1": visits,
    "Metric2": percentages_data["Saw a Result"],
    "Metric3": percentages_data["1st common reason (Unknown)"],
    "Metric4": percentages_data["2nd common reason (Disconnected)"],
}

df = pd.DataFrame(data)

# Calculate the average of "Saw a Result" for the last 7 days (excluding the latest date)
average_saw_a_result = sum(percentages_data["Saw a Result"][:-1]) / 7

# Add the average as a new column
df["Last 7 days"] = average_saw_a_result

# Display the DataFrame in the desired format
df_output = df[['Dates', 'Operating System Type', 'Metric1', 'Metric2', 'Metric3', 'Metric4', 'Last 7 days']]
print(df_output)
```
